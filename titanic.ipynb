```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
data = sns.load_dataset("titanic")
data.head()
```

```python
# drop deck, embark_town,
data.drop(["deck","embark_town","alive","class","who","adult_male"],axis=1,inplace=True)
```

```python
data.head()
```

```python
data.describe()
```

```python
# fill null ages with mean
data["age"]=data["age"].fillna(data["age"].mean(),inplace=True)
```

```python
data.describe()
```

```python
data.info()
```

```python
# instead of filling those 2 rows in embarked, delete those 2 null rows
data.dropna(subset=["embarked"],inplace=True)
```

```python
data.info()
```

```python
# we have to predict survived column
# label encode
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()

# label encode sex and embarked
data["sex"] = enc.fit_transform(data["sex"])
data["embarked"] = enc.fit_transform(data["embarked"])
```

```python
data.head()
```

```python
data = data.astype(int)
```

```python
data.head()
```

```python
data["pclass"].value_counts()
```

```python
# in X we keep input, in Y output
X = data.drop("survived",axis = 1)
Y = data["survived"]
```

```python
# train test split
from sklearn.model_selection import train_test_split

X_Train, X_Test, y_train, y_test = train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X=X_Train,y=y_train)
```

```python
y_test
```

```python
y_pred = model.predict(X=X_Test)
print(y_pred)
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

print("Confusion Matrix: ")
print(confusion_matrix(y_true=y_test, y_pred=y_pred))

print("Accuracy Score: ")
print(accuracy_score(y_true=y_test, y_pred=y_pred))

print("Report: ")
print(classification_report(y_true=y_test, y_pred=y_pred))
```

```python
from sklearn.metrics import recall_score, f1_score, accuracy_score, precision_score

accuraccy = accuracy_score(y_true=y_test,y_pred=y_pred)
# print("accuraccy: ", accuraccy)
print("accuraccy: ", accuraccy*100)

precision = precision_score(y_true=y_test,y_pred=y_pred)
# print("precision: ", precision)
print("precision: ", precision*100)

recall = recall_score(y_true=y_test,y_pred=y_pred)
# print("recall: ", recall)
print("recall: ", recall*100)

f1 = f1_score(y_true=y_test, y_pred=y_pred)
# print("F1 scrore: ", f1)
print("F1 scrore: ", f1*100)
```
