```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")
```

```python
data = sns.load_dataset("iris")

data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
# data['species'].value_counts()
data['species'].unique()
```

```python
# train test split
from sklearn.model_selection import train_test_split

X = data.drop("species", axis=1)
y = data["species"]

X_train,X_test, y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)

# use knn model
from sklearn.neighbors import KNeighborsClassifier

# trying out different values of n_neighbors will affect model accuracy
model_knn = KNeighborsClassifier(n_neighbors=51)
model_knn.fit(X_train,y_train)

model_knn.score(X_test,y_test)
```

```python
# SVM
# try different parameters with this too

from sklearn.svm import SVC

model_svm = SVC(gamma="auto")

model_svm.fit(X_train,y_train)

model_svm.score(X_test,y_test)
```

### grid search cv

```python
from sklearn.model_selection import GridSearchCV
# in grid search, no need to do train test split

# , C=30 ,kernel="rbf"
# param_grid dictonary
param_grid = {
    "C":[1,10,20,30],
    "kernel":["linear","poly","rbf"]
}

classifier = GridSearchCV(model_svm,param_grid,cv=5, return_train_score=False)

classifier.fit(X,y)
```

```python
# classifier.score(X_test,y_test)

classifier.cv_results_
```

```python
result = pd.DataFrame(classifier.cv_results_)
result
```

```python
res = result[["param_C","param_kernel","mean_test_score","rank_test_score"]]
# res.sort_values("rank_test_score")
res.sort_values("mean_test_score")
res
```
